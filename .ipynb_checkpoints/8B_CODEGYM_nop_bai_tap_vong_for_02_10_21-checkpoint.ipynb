```python
# Bài Fizz Buzz: nhập hai số, sử dụng vòng for để trả về ( 02_10_21 )
# in ra các số trong khoảng này, 
# nếu số đó chia hết cho 3, in ra “Fizz” thay vì nó, 
# nếu số đó chia hết cho 5, in ra “Buzz” 
# và nếu số đó chia hết cho cả 3 và 5, hãy in ra “FizzBuzz”.

n1 = int(input("nhập số thứ nhất :"))
n2 = int(input("nhập số thứ hai :"))
print("""Bài Fizz Buzz: nhập hai số, sử dụng vòng for để trả về
----------------------------------------------------------""")
for i in range(n1,n2+1):
    if i % 3 == 0 and i % 5 != 0:
        print("Fizz")
    elif i % 5 == 0 and i % 3 !=0:
        print("Buzz")
    elif i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    else:
        print(i)
        
```

```python
# Bài sinh bảng cửu chương: hiển thị tất cả các bảng cửu chương từ 1 đến 9 ( 02_10_21 )
print("""Hiển thị tất cả các bảng cửu chương từ 1 đến 9
---------------------------------------------------""")
for i in range(2,10):
    for j in range(1, 11):
        tich = i * j
        print(i, "x", j, "=", tich)
```

```python
# Nhập vào hai số, tách sổ, lấy số đầu, số cuối, so sánh
print("""Bài Fizz Buzz: nhập số có hai chữ số, sử dụng vòng for để trả về
----------------------------------------------------------""")
# Nhập số đầu vào có hai chữ số

n = input("nhập vào một số có hai chữ số: ")

# Tách lấy số đầu, số cuối
so1 = int(n[0])
so2 = int(n[1])

# Kiểm tra biến nhập, so sánh hai số:
if so1 > so2: 
    print("số kết thúc cần lớn hơn số bắt đầu")
else:
    print("""bắt đầu xử lý việc in kết quả
-----------------------------""")

# Bắt đầu xử lý việc in kết quả

for i in range(so1,so2+1):
    if i % 3 == 0 and i % 5 != 0:
        print("Fizz")
    elif i % 5 == 0 and i % 3 !=0:
        print("Buzz")
    elif i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    else:
        print(i)


```
